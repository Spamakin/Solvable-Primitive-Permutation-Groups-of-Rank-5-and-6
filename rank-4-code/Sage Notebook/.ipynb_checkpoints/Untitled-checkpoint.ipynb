```python
maxims = []
for i in range(74):
    G = gap.PrimitiveGroup(64,i+1)
    if gap.IsSolvableGroup(G):
        G0 = gap.Stabilizer(G,1)
        if len(gap.Orbits(G0)) < 4:
            maxims.append(G)
for m in maxims:
    for n in maxims:
        if gap.IsSubgroup(m,n):
            maxims.remove(n)
for m in maxims:
    m0 = gap.Stabilizer(m,1)
    if len(gap.Orbits(m0)) == 3:
        print(gap.Order(m))
```
